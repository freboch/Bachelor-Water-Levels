```python
import pickle
import numpy as np
import torch
import matplotlib.pyplot as plt

from prep_lake_data import *
from eval_model_plots import plot_data
from project_func import proj_to_sphere
```

```python
# Get original lake data dictionary from "data_dict_build.ipynb"
with open(f'saves/lake_data_dicts.pkl', 'rb') as f:
            data = pickle.load(f)
```

```python
def all_wl_obs(data_df):
    # Get all wlts values
    wltss = {'conv':{'all_wltsv':[],'all_wlts':[], 'all_wlts_lakes':[], 'all_wltsv_t': [], 'all_wlts_t': [], 'wlts_std': [], 'lakes': []},
            'all':{'all_wltsv':[],'all_wlts':[], 'all_wlts_lakes':[], 'all_wltsv_t': [], 'all_wlts_t': [], 'wlts_std': [], 'lakes': []},
            'nn':{'all_wltsv':[],'all_wlts':[], 'all_wlts_lakes':[], 'all_wltsv_t': [], 'all_wlts_t': [], 'wlts_std': [], 'lakes': []}}

    for lake in data_df.keys():
        for i, df in enumerate(data_df[lake]['wlts']):
            wl  =list(df.wl)
            wltss['all']['all_wltsv'].extend(wl)
            wltss['all']['all_wlts'].append(wl)
            wltss['all']['wlts_std'].append(np.std(wl))
            wltss['all']['all_wlts_lakes'].append(lake)
            wltss['all']['all_wltsv_t'].extend(list(df.time))
            wltss['all']['all_wlts_t'].append(list(df.time))
            if lake not in wltss['all']['lakes']:
                wltss['all']['lakes'].append(lake)

            if data_df[lake]['conv'][i].conv[0]==0:
                wltss['conv']['all_wltsv'].extend(wl)
                wltss['conv']['all_wlts'].append(wl)
                wltss['conv']['wlts_std'].append(np.std(wl))
                wltss['conv']['all_wlts_lakes'].append(lake)
                wltss['conv']['all_wltsv_t'].extend(list(df.time))
                wltss['conv']['all_wlts_t'].append(list(df.time))
                if lake not in wltss['conv']['lakes']:
                    wltss['conv']['lakes'].append(lake)
            if 'nn' in data_df[lake].keys() and data_df[lake]['nn'][i]:
                wltss['nn']['all_wltsv'].extend(wl)
                wltss['nn']['all_wlts'].append(wl)
                wltss['nn']['wlts_std'].append(np.std(wl))
                wltss['nn']['all_wlts_lakes'].append(lake)
                wltss['nn']['all_wltsv_t'].extend(list(df.time))
                wltss['nn']['all_wlts_t'].append(list(df.time))
                if lake not in wltss['nn']['lakes']:
                    wltss['nn']['lakes'].append(lake)
    return wltss
```

# raw lake data

```python
wltss = all_wl_obs(data)
```

```python
fig, ax = plt.subplots(1, 1,figsize=(8, 5))
for i, wlts in enumerate(wltss['all']['all_wlts']):
    x = wltss['all']['all_wlts_t'][i]
    y = wlts
    ax.plot(x,y)
ax.set_xlabel('time (decimal year)')
ax.set_ylabel('water level (m)')
ax.set_title(f'Data: original lakes')
plt.savefig(f"plots/original lakes.png", 
                    bbox_inches='tight', facecolor ="w", dpi=200)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1,figsize=(8, 5))
for i, wlts in enumerate(wltss['all']['all_wlts']):
    x = wltss['all']['all_wlts_t'][i]
    y = proj_to_sphere(torch.tensor([wlts])).T
    ax.plot(x,y)
ax.set_xlabel('time (decimal year)')
ax.set_ylabel('water level (m)')
ax.set_title(f'Data: original lakes, projected')
plt.savefig(f"plots/original lakes projected.png", 
                    bbox_inches='tight', facecolor ="w", dpi=200)
plt.show()
```
